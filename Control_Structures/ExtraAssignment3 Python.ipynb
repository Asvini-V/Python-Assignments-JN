```python
# Catagory the people by their age like children, adult, citizen, senior

age=int(input("Age:"))
if(age<18):
    print("Children")
elif(age<35):
    print("Adult")
elif(age<59):
    print("Citizen")
else:
    print("Senior Citizen")
```

```python
# Find whether given number is positive or negative

numb=int(input("Enter any number:"))
if (numb>0):
    print("No is positive")
elif (numb<0):
    print("No is negative")
else:
    print("No is zero")
```
